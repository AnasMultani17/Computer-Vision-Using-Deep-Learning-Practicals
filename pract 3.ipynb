```python
import cv2
import os
from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt
```

```python
OI=cv2.imread('temp.jpg')
cv2_imshow(OI)
```

```python
def average(image,size,kernel):
  out=image
  h,w,c=out.shape
  sz=size//2
  for i in range(sz,(h-sz)):
    for j in range(sz,(w-sz)):
      for k in range(c):
        area = image[i-sz:i+sz+1, j-sz:j+sz+1, k]
        sum=np.sum(kernel)
        out[i-sz, j-sz, k] = np.sum(area * kernel/sum)
  return out
```

```python
kernel=np.array([[1,1,1],[1,1,1],[1,1,1]])
DI=average(OI,3,kernel)
cv2_imshow(DI)
kernel1 = np.array([
    [1,  4,  6,  4, 1],
    [4, 16, 24, 16, 4],
    [6, 24, 36, 24, 6],
    [4, 16, 24, 16, 4],
    [1,  4,  6,  4, 1]
])
DI1=average(OI,5,kernel1)
cv2_imshow(DI1)
```

```python
def mean(image, size):
    h, w, c = image.shape
    sz = size // 2
    ans = np.zeros_like(image, dtype=np.float32)
    for i in range(sz, h - sz):
        for j in range(sz, w - sz):
            for k in range(c):
                area = image[i-sz:i+sz+1, j-sz:j+sz+1, k]
                ans[i, j, k] = np.mean(area)
    return ans
```

```python
DI2=mean(OI,3)
cv2_imshow(DI2)
```

```python
def med(image, size):
    h, w, c = image.shape
    sz = size // 2
    ans = np.zeros_like(image, dtype=np.float32)
    for i in range(sz, h - sz):
        for j in range(sz, w - sz):
            for k in range(c):
                area = image[i-sz:i+sz+1, j-sz:j+sz+1, k]
                ans[i, j, k] = np.median(area)
    return ans
```

```python
DI3=med(OI,5)
cv2_imshow(DI3)
```

```python
def max(image, size):
    h, w, c = image.shape
    sz = size // 2
    ans = np.zeros_like(image, dtype=np.float32)
    for i in range(sz, h - sz):
        for j in range(sz, w - sz):
            for k in range(c):
                area = image[i-sz:i+sz+1, j-sz:j+sz+1, k]
                ans[i, j, k] = np.max(area)
    return ans
```

```python
def min(image, size):
    h, w, c = image.shape
    sz = size // 2
    ans = np.zeros_like(image, dtype=np.float32)
    for i in range(sz, h - sz):
        for j in range(sz, w - sz):
            for k in range(c):
                area = image[i-sz:i+sz+1, j-sz:j+sz+1, k]
                ans[i, j, k] = np.min(area)
    return ans
```

```python
def laplatian(image,size,kernel):
  out=image
  h,w,c=out.shape
  sz=size//2
  for i in range(sz,(h-sz)):
    for j in range(sz,(w-sz)):
      for k in range(c):
        area = image[i-sz:i+sz+1, j-sz:j+sz+1, k]
        out[i-sz, j-sz, k] = np.sum(area * kernel)
  return out
```

```python
k=np.array([[0,1,0],[1,-4,1],[0,1,0]])
DI5=laplatian(OI,3,k)
cv2_imshow(DI5)
```

```python
def robert(image, k1):
    h, w, c = image.shape
    out = np.zeros_like(image)

    for i in range(h - 1):
        for j in range(w - 1):
            for k in range(c):
                area = image[i:i+2, j:j+2, k]
                out[i, j, k] = np.sum(area * k1)
    return out
```

```python
k=np.array([[-1,0],[0,1]])
DI6=robert(OI,k)
cv2_imshow(DI6)
```

```python
def sobel(image,size,ke):
  out=image
  h,w,c=out.shape
  sz=size//2
  for i in range(sz,(h-sz)):
    for j in range(sz,(w-sz)):
      for k in range(c):
        area = image[i-sz:i+sz+1, j-sz:j+sz+1, k]
        out[i-sz, j-sz, k] = np.sum(area * ke)
  return out
```
